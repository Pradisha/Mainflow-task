```python
#Getting ans of 1st question

#1 How many people have heart disease and how many people doesnt have heart disease?
#getting the values
df.target.value_counts()
```

```python
# 0- Female, 1- Male
# 0- No disease, 1- Disease
#counting male and female
df.sex.value_counts()
```

```python
#2nd ques
#2 People of which sex has the most heart disease
pd.crosstab(df.target, df.sex)
```

```python
palette = {1: 'royalblue', 0: 'deeppink'}
sns.countplot(x = 'target', data=df, hue = 'sex' , palette = palette)
plt.title("Heart disease frequency vs Gender")
plt.xlabel("0 = No heart disease, 1 = Heart Disease");
plt.legend(title='Sex', loc='upper right', labels=['Female', 'Male'])
```

```python
#5th ques
#how does cholestrol level vary with gender?

# Create a new column for cholesterol level ranges
df['chol_range'] = pd.cut(df['chol'], bins=[0, 200, 240, 500], labels=['<200', '200-240', '>240'])

# Set the plot size
plt.figure(figsize=(10, 6))

# Create the count plot
sns.countplot(x='chol_range', hue='sex', data=df, palette={0: 'pink', 1: 'blue'})

# Add titles and labels
plt.title('Count of Cholesterol Levels by Gender')
plt.xlabel('Cholesterol Level Range')
plt.ylabel('Count')
plt.legend(title='Sex', loc='upper left', labels=['Female', 'Male'])

# Display the plot
plt.show()
```

```python
#we can conclude that males have high cholestrol level
```

```python
#ques 6 Which age group has the most cholestrol level
```

```python
# Create a new column for cholesterol level ranges
df['chol_range'] = pd.cut(df['chol'], bins=[0, 200, 240, 500], labels=['<200', '200-240', '>240'])

# Create a new column for age groups
df['age_group'] = pd.cut(df['age'], bins=[20, 30, 40, 50, 60, 70, 80], labels=['20-29', '30-39', '40-49', '50-59', '60-69', '70-79'])

# Set the plot size
plt.figure(figsize=(14, 8))

# Create the count plot
sns.countplot(x='age_group', hue='chol_range', data=df, palette='viridis')

# Add titles and labels
plt.title('Count of Age Groups by Cholesterol Level Ranges')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Cholesterol Level Range', loc='upper left')

# Display the plot
plt.show()
```

```python
#we can conclude that people in th age group 50-60 has the most cholestrol level
#Ques 7 which age group has lowest blood sugar level?
```

```python
# Create a new column for cholesterol level ranges
df['age_group'] = pd.cut(df['age'], bins=[20, 30, 40, 50, 60, 70, 80], labels=['20-29', '30-39', '40-49', '50-59', '60-69', '70-79'])

# Set the plot size
plt.figure(figsize=(10, 6))

# Create the count plot
sns.countplot(x='age_group', hue='fbs', data=df, palette={0: 'lightcoral', 1: 'teal'})

# Add titles and labels
plt.title('Count of Age groups by fasting blood sugar levels')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Blood sugar level>120 mg/dl', loc='upper right', labels=['False', 'True'])

# Display the plot
plt.show()
```

```python
#Conclusion 
#Age group 20-29 doesnot have fbs>120mg/dl
```
