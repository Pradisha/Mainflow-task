```python
# Number of movies and TV shows
content_type_distribution = df['type'].value_counts()
palette = {'Movie': 'royalblue', 'TV Show': 'thistle'}
# Plot content type distribution
sns.countplot(data=df, x='type', palette = palette)
plt.title('Disney+ Content Type Distribution')
```

```python
# Distribution of content by release year
release_year_distribution = df.head(10)['release_year'].value_counts().sort_index()
# Plot release year distribution

release_year_distribution.plot(kind='bar', color ='orchid')
plt.title('Release Year Distribution of top 10 contents')
```

```python
# Titles added each year
titles_added_each_year = df['date_added'].dt.year.value_counts().sort_index()
# Plot titles added each year
titles_added_each_year.plot(kind='bar', color ='mediumpurple')
plt.title('Titles Added Each Year')
```

```python
# Most common genres

#Splits each genre string into a list of genres, using , as the delimiter.
genres = df['listed_in'].str.split(', ')

#list comprehension that iterates over each sublist in genres and 
#then over each genre in the sublist, effectively flattening the list of lists into 
#a single list containing all genres
all_genres = [genre for sublist in genres.dropna() for genre in sublist]

#Converts the list of all genres into a pandas Series.
genre_distribution = pd.Series(all_genres).value_counts()

# Plot genre distribution
genre_distribution.head(10).plot(kind='pie')
plt.ylabel('')
plt.title('Top Genres on Disney+')
```

```python
# Most prolific directors
most_prolific_directors = df['director'].value_counts().head(10)

#Plot most prolific directors
most_prolific_directors.plot(kind='bar', color='lightseagreen')
plt.title('Top 10 Prolific Directors')
```

```python
# Most frequent cast members
cast_members = df['cast'].str.split(', ')
all_cast_members = [cast for sublist in cast_members.dropna() for cast in sublist]
most_frequent_cast_members = pd.Series(all_cast_members).value_counts().head(10)
#plotting frequent casts
most_frequent_cast_members.plot(kind='bar', color='greenyellow')
plt.title(' Most frequent cast members')
plt.xlabel('Casts')
```

```python
# Distribution of content by rating
rating_distribution = df['rating'].value_counts()
rating_distribution.plot(kind='bar')
plt.title(' distribution of Ratings')
plt.ylabel('Count')
```

```python
# Type of content by country
content_by_country = df.groupby('country')['type'].value_counts().unstack().fillna(0).head(10)
content_by_country.plot(kind='bar', color=['orchid','cyan'])
plt.title(' Type of content by country')
plt.ylabel('Count')
```

```python
# Distribution of movie durations
movie_durations = df.head(50)[df['type'] == 'Movie']['duration'].str.replace(' min', '').astype(int)
plt.figure(figsize=(10, 6))
plt.hist(movie_durations, bins=30, color='goldenrod', edgecolor='purple')
plt.title('Distribution of Top 50 Movie Durations on Disney+')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```
