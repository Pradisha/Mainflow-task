 _Importing Libraries_

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

Loading Dataset

```python
data = pd.read_csv("C:\\Users\\Asus\\Documents\\Flipkart_project\\Dataset-SA.csv")
```

Viewing the top rows of the dataframe

```python
data.head()
```

```python
#Concise summary of data
data.info()
```

_Data Cleaning_

```python
#Checking for null values 
data.isnull().sum()
```

```python
#dropping rows with null values
data1 = data.dropna()
data1.isnull().sum()
```

```python
#Checking the dimensionality of the dataframe
data1.shape
```

```python
#Removing the unnecessary duplicates
data1 = data1.drop_duplicates()
data1.shape
```

```python
#Removing Non numeric values from Rate Column
data1['Rate']= pd.to_numeric(data1['Rate'], errors='coerce')
data1.dropna(subset = ['Rate'], inplace = True)
data1.head()
```

Lets calculate and visualize the rating distribution

```python
#Rating distribution
rating_sum_up= data1['Rate'].value_counts()
```

```python
#Visualizing/Plotting the rating distribution
plt.figure(figsize=(8,6))
rating_sum_up.plot(kind = 'bar', color ='Teal')
plt.title('Distribution of Rating')
plt.xlabel('----> Rating')
plt.xticks(rotation= 0)
plt.ylabel('Review Count')
#Displaying 
plt.show()
```

```python
#Lets filter Positive Reviews(Specifically 4 star and 5 star)
pos_reviews = data1[data1['Rate']>= 4]
#Genarating wordcloud for positive reviews
pos_txt = ' '.join(pos_reviews['Review']) #To join all into one string
pos_wordcloud = WordCloud(width= 800, height = 600, background_color = 'white', colormap= 'viridis').generate(pos_txt)

#Displaying the wordcloud
plt.figure(figsize =(10,6))
plt.imshow(pos_wordcloud, interpolation='bilinear') #for smoother text appearance
plt.axis('off')
plt.title('Positive Reviews Word Cloud')
plt.show()
```

```python
#Now filter negative reviews(1 star and 2 star)
neg_reviews = data1[data1['Rate']<= 2]
#Genarating wordcloud for negative reviews
neg_txt = ' '.join(neg_reviews['Review']) #To join all into one string
neg_wordcloud = WordCloud(width= 800, height = 600, background_color = 'white', colormap= 'copper').generate(neg_txt)

#Displaying the wordcloud
plt.figure(figsize =(10,6))
plt.imshow(neg_wordcloud, interpolation='bilinear') #for smoother text appearance
plt.axis('off')
plt.title('Negative Reviews Word Cloud')
plt.show()
```

```python
#Calculating Average Rating of each product decribed
```

```python
prod_rating = data1.groupby('product_name')['Rate'].mean().sort_values()
prod_rating
```

```python
prod_ratings = data1.groupby('product_name')['Rate'].mean().sort_values()

# Specify the number of top products to visualize
top = 7  

# Select the top N products by average rating
top_prod_ratings = prod_ratings.tail(top)

# Plot the average rating for the top N products
plt.figure(figsize=(15, 4))
top_prod_ratings.plot(kind='bar', color='orchid')
plt.xlabel('Average Rating')
plt.ylabel('Product ')
plt.title(f'Average Rating for Top {top} Products')
plt.show()
```

```python
prod_ratings = data1.groupby('product_name')['Rate'].mean().sort_values()

# Specify the number of top products to visualize
bottom = 7  

# Select the top N products by average rating
bottom_prod_ratings = prod_ratings.head(bottom)

# Plot the average rating for the top N products
plt.figure(figsize=(15, 4))
bottom_prod_ratings.plot(kind='bar', color='salmon')
plt.xlabel('Average Rating')
plt.ylabel('Product ')
plt.title(f'Average Rating for bottom {bottom} Products')
plt.show()
```

```python
combined_prod_ratings = pd.concat([top_prod_ratings, bottom_prod_ratings])
plt.figure(figsize=(12, 10))
combined_prod_ratings.plot(kind='bar', color='goldenrod')
plt.xlabel('----> Products')
plt.ylabel('Ratings')
plt.title(f'Average Rating for {bottom} Top and Bottom  Products')
plt.show()
```

```python
#The sentiments are in string values
#Mappping them into numeric values for easy Visualization
sentiment_mapping = {'positive':1 , 'neutral':0 , 'negative':-1}
data1['Sentiment_value'] = data1['Sentiment'].map(sentiment_mapping)
print(data1.head())
```

```python
print(data1['Sentiment_value'].value_counts())
```

```python
# Calculate sentiment distribution
sentiment_counts = data1['Sentiment'].value_counts()

# Plot sentiment distribution
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['lightseagreen', 'firebrick', 'goldenrod'])
plt.xlabel('---> Sentiments')
plt.xticks(rotation=0)
plt.ylabel('Number of Reviews')
plt.title('Sentiment Distribution')
plt.show()
```

```python
pd.crosstab(data1.Rate.head(8000), data1.Sentiment.head(8000)).plot(kind = 'bar', color = ['firebrick', 'goldenrod','lightseagreen'])
plt.title('Rating vs Sentiment')
plt.xticks(rotation=50)
plt.xlabel('---> Rating');
plt.ylabel('Count')
```

```python
#Handling invalid entries in Product price column
data1['product_price']= pd.to_numeric(data1['product_price'], errors='coerce')
#Dropping Nan values
data1.dropna(subset = ['product_price'], inplace = True)
data1.head()
```

```python
# Create a new column for Product price ranges
data1['price_range'] = pd.cut(data1['product_price'], bins=[0, 2000, 7000, 11000, 18000], labels=['<2000', '2000-7000', '70001-11000','>11000'])

# Set the plot size
plt.figure(figsize=(10, 6))

# Create the count plot
sns.countplot(x='price_range', hue='Rate', data=data1, palette={1: 'firebrick',2:'gold',3:'skyblue',4:'hotpink',5:'forestgreen'})

# Add titles and labels
plt.title('Price vs Rating')
plt.xlabel('Price_Range')
plt.ylabel('Count')

# Display the plot
plt.show()
```

We can conclude that Products listed with amount lesser than 2000 has the highest 5 star rating

```python
# Create a new column for Product price ranges
data1['price_range'] = pd.cut(data1['product_price'], bins=[0, 2000, 7000, 11000, 18000], labels=['<2000', '2000-7000', '70001-11000','>11000'])

# Set the plot size
plt.figure(figsize=(10, 6))

# Create the count plot
sns.countplot(x='price_range', hue='Sentiment', data=data1, palette={'lightseagreen', 'firebrick', 'goldenrod'})

# Add titles and labels
plt.title('Price vs Rating')
plt.xlabel('Price_Range')
plt.ylabel('Count')

# Display the plot
plt.show()
```
